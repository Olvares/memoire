## Modèle en cas de non imigration

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

\def\dx{\mathrm{d}}
\newcommand{\dxdt}[2][t]{\frac{\dx#2}{\dx#1}}
\begin{equation}
    \left\{\begin{array}{rl}
        \dxdt{x_1}	& =	- \alpha x_1 y - \lambda x_1	\\
        \dxdt{y}	& =	e_1\alpha x_1 y + \gamma w -\delta y - \mu_y y	\\
        \dxdt{z}	& =	\delta y - \mu_z z	\\
        \dxdt{w}	& =	\rho z - \gamma w - \mu_w w	\\
    \end{array}\right.
    \label{eq1}
\end{equation}

```python
def pop_dyn_nim(x, y, z, w, an, l, ga, d, rh, my, mz, mw, e):
    dx = -(an*y + l) * x
    dy = e*an*x*y + ga*w - (d+my) * y
    dz = d*y - mz*z
    dw = rh*z - (ga + mw) * w

    return np.array([dx, dy, dz, dw])
```

```python
def RK4(f, I, y0, n):
    h = (I[1] - I[0])/n
    t = np.linspace(I[0], I[1],n+1)
    m = y0.size
    y = np.zeros((n+1, m))
    y[0,:] = y0
    for i in range(n):
        k1 = h * f(t[i], y[i,:])
        k2 = h * f(t[i] + h/2, y[i,:] + k1/2)
        k3 = h * f(t[i] + h/2, y[i,:] + k2/2)
        k4 = h * f(t[i] + h, y[i,:] + k3)

        y[i+1,:] = y[i,:] + (k1 + 2 * (k2 + k3) + k4) / 6
    return  t,y
```

## A litle nverification of RK4
