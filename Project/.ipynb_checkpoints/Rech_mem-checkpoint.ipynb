# Daudi's model
## Expression in non immigration case

\begin{equation}
    \left\{\begin{array}{l}
        \frac{\mathrm{d}x_1}{\mathrm{d}t} =	- \alpha x_1 y - \lambda x_1	\\
        \frac{\mathrm{d}y}{\mathrm{d}t} =	e_1\alpha x_1 y + \gamma w -\delta y - \mu_y y	\\
        \frac{\mathrm{d}z}{\mathrm{d}t} =	\delta y - \mu_z z	\\
        \frac{\mathrm{d}w}{\mathrm{d}t} =	\rho z - \gamma w - \mu_w w	\\
    \end{array}\right.
    \label{eq1}
\end{equation}

## Simulation of Daudi's model using RK4
### initialisation

```python
k = 500
y0 = 0
z0 = np.array([15, 30, 45, 60])
w0 = 0

r = 0.0417
d = 0.071
ga = 0.071
my = 0.0071
mz = 0.115
mw = 0.04
l = 0.015
e = 1.6
an = 0.000154

rate = np.array([an, l, ga, d, r, my, mz, mw, e])

t0 = 0
t1 = 63
T = 160

I1 = [t0,T]
leg = ['Maize', 'Caterpillar', 'Adult', 'Eggs']
```

### Simulation

```python
i = 0
s0 = np.array([k, y0, z0[i], w0])
solD = myM.solver(rate, s0, I1)

plt.figure(1)
for i in range(z0.size):
    s0 = np.array([k, y0, z0[i], w0])
    solD.set_input(rate, s0, I1)
    solD.daudiModSol()
    solD.printSol(plt)
    
fig, axs = plt.subplots(2,2)
for i in range(z0.size):
    s0 = np.array([k, y0, z0[i], w0])
    solD.set_input(rate, s0, I1)
    solD.daudiModSol()
    solD.printSol(axs)

#plt.show()
```
